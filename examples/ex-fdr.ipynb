```python
# FRED: 주간 실업수당 청구 건수(ICSA)
# 경기가 나빠지는 경우 후행지표로 동작, 경기가 좋아지는 경우 선행지표로 동작
df = fdr.DataReader('FRED:NASDAQCOM,ICSA', start='2006', end='2010-10-30')
df.plot(secondary_y='NASDAQCOM')
```

```python
# 2008년 금융위기 나스닥지수(NASDAQCOM)와 주간 실업수당 청구 건수(ICSA)

df = fdr.DataReader('FRED:NASDAQCOM,ICSA', start='2006', end='2010')
ax = df.plot(secondary_y='NASDAQCOM')
ax = ax.axvspan('2007-12-01', '2009-03-30', color='gray', alpha=0.2)
```

```python
# 최근(2020년 10월), 나스닥지수(NASDAQCOM)와 주간 실업수당 청구 건수(ICSA)

df = fdr.DataReader('FRED:NASDAQCOM,ICSA', start='2020-01-01', end='2024-06-30')
ax = df.plot(secondary_y='NASDAQCOM')
```

```python
# FRED: 소비자심리지수 (UMCSENT)
# 100을 기준으로 100이상이면 긍정적 100이하면 부정적

# 2008년 금융위기 나스닥지수(NASDAQCOM)와 미시건대 소비자심리지수(UMCSENT)

df = fdr.DataReader('FRED:NASDAQCOM,UMCSENT', start='2020', end='2024')
ax = df.plot(secondary_y='NASDAQCOM')
# ax = ax.axvspan('2007-12-01', '2009-03-30', color='gray', alpha=0.2)
```

```python
# FRED: 주택 판매 지수 (HSN1F)
# 주택시장 활성화 정도
# 불황으로 주가가 하락할 때 하락에 앞서 주택 판매 지수 하락
# 2008년 금융위기 나스닥지수(NASDAQCOM)와 주택 판매 지수

df = fdr.DataReader('FRED:NASDAQCOM,HSN1F', start='2005', end='2010-10-30')
ax = df.plot(secondary_y='NASDAQCOM')
ax = ax.axvspan('2007-12-01', '2009-03-30', color='gray', alpha=0.2)
```

```python
# 최근(2020년 10월), 나스닥지수(NASDAQCOM)와 주택판매지수(HSN1F)

df = fdr.DataReader('FRED:NASDAQCOM,HSN1F', start='2019', end='2010-10-30')
ax = df.plot(secondary_y='NASDAQCOM')
```

```python
# FRED: 실업률 (UNRATE)
# 불황에는 후행지표, 경기회복에 동행지표
# 2008년 금융위기 나스닥지수(NASDAQCOM)와 실업률(UNRATE)

df = fdr.DataReader('FRED:NASDAQCOM,UNRATE', start='2005', end='2010-10-30')
ax = df.plot(secondary_y='NASDAQCOM')
ax = ax.axvspan('2007-12-01', '2009-03-30', color='gray', alpha=0.2)
```

```python
# FRED: M2 통화량 (M2SL)
# 시중 돈이 얼마나 많이 풀렸는지(유동성)을 보여주는 지표
# 2008년 금융위기, 나스닥지수(NASDAQCOM)와 M2통화량(M2SL)

df = fdr.DataReader('FRED:NASDAQCOM,M2SL', start='2005', end='2010-10-30')
ax = df.plot(secondary_y='NASDAQCOM')
ax = ax.axvspan('2007-12-01', '2009-03-30', color='gray', alpha=0.2)
```

```python
# FRED: 하이일드 채권 스프레드 (BAMLHOAOHYM2)
# 낮은 신용등급 회사에 투자하는 고위험,고수익 채권
# 스프레드는 국채 수익률과의 차이
# 하이일드 채권 스프레드 = 하이일드 채권 수익률 – 국채 수익률
# 불황일 수록 하이일드 채권 스프레드는 더 커짐
# 2008년 금융위기, 나스닥지수(NASDAQCOM)와 하이일드 채권 스프레드

df = fdr.DataReader('FRED:NASDAQCOM,BAMLH0A0HYM2', start='2005', end='2010-10-30')
ax = df.plot(secondary_y='NASDAQCOM')
ax = ax.axvspan('2007-12-01', '2009-03-30', color='gray', alpha=0.2)
```
